```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import time
```

```python
# read in csv file
weather_df = pd.read_csv('cities.csv')
```

```python
# set index to City_ID
weather_df.set_index('City_ID', inplace=True)

# preview
weather_df.head()
```

## Plotting the Data
* Use proper labeling of the plots using plot titles (including date of analysis) and axes labels.
* Save the plotted figures as .pngs.

## Temperature vs. Latitude Plot

```python
# Create a Scatter Plot for temperature vs latitude
x_values = weather_df['Lat']
y_values = weather_df['Max Temp']
plt.scatter(x_values,y_values)
plt.xlabel('Latitude')
plt.ylabel('Temperature (F)')
plt.title('Latitude vs Temperature (F) (04/20/22)')
plt.xlim(-85,85)
plt.savefig("../assets/images/01_temp_vs_lat.png")
plt.show()
```

## Humidity vs. Latitude Plot

```python
# Create a Scatter Plot for latitude vs humidity
x_values = weather_df['Lat']
y_values = weather_df['Humidity']
plt.scatter(x_values,y_values)
plt.xlabel('Latitude')
plt.ylabel('Humidity (%)')
plt.title('Latitude vs Humidity (%) (04/20/22)')
plt.xlim(-85,85)
plt.savefig("../assets/images/02_hum_vs_lat.png")
plt.show()
```
